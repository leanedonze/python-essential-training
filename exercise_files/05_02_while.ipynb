```python
from datetime import datetime
```

## While loops

```python
datetime.now().second
```

```python
wait_until = datetime.now().second + 2

while datetime.now().second != wait_until:
    print('Still waiting!')
    
print(f'We are at {wait_until} seconds!')
```

```python
wait_until = datetime.now().second + 2

while datetime.now().second != wait_until:
    1 + 1
    
print(f'We are at {wait_until} seconds!')
```

### Pass

```python
wait_until = datetime.now().second + 2

while datetime.now().second != wait_until:
    # Do nothing, or use when a function or something is not implemented yet
    pass
    
print(f'We are at {wait_until} seconds!')
```

### Break

```python
wait_until = datetime.now().second + 2

while True:
    if datetime.now().second == wait_until:
        print(f'We are at {wait_until} seconds!')
        # Break statement to go out of the loop
        break
```

```python
# !Be careful, break statement go out of one loop (inner one)
wait_until = datetime.now().second + 2

while True:
    while datetime.now().second == wait_until:
        print(f'We are at {wait_until} seconds!')
        break
```

### Continue

```python
# Continue statement to skip the lines after
wait_until = datetime.now().second + 2

while datetime.now().second != wait_until:
    continue
    print('Still waiting!')
    
    
print(f'We are at {wait_until} seconds!')
```

```python
wait_until = datetime.now().second + 2

while True:
    if datetime.now().second < wait_until:
        continue
    break
    
    
print(f'We are at {wait_until} seconds!')
```
